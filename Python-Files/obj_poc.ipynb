```python
import tkinter as tk
import math
```

```python
#parse obj

cube_file = open('Car.obj','r')
cube_file_lines = cube_file.readlines()

vertices = []
normals = []
faces = []

for line in cube_file_lines:
    split = line.split()
    if len(split) > 0:
        if split[0] == 'v':
            vertices.append(split[1:4])
        if split[0] == 'vn':
            normals.append(split[1:4])
        if split[0] == 'f':
            faces.append(split[1:4])

float_vertices = []
for vertex in vertices:
    coord = []
    for num in vertex:
        coord.append(float(num))
    float_vertices.append(coord)
vertices = float_vertices
```

```python
def getFaceCoords(face, vertex_lookup):
    coords = []
    for i in range(3):
        vertices = face[i].split('/')
        vertex_coords = vertex_lookup[int(vertices[0])-1]
        coords.append(vertex_coords)
    return coords

def adjustCoordsToScreen(vertices, screen_x, screen_y, screen_z):
    new = []
    for vertex in vertices:
        coord = []
        for i in range(3):
            if i == 0:
                coord.append(float(vertex[i]) + SCREEN_X)
            if i == 1:
                coord.append(float(vertex[i]) + SCREEN_Y)
            if i == 2:
                coord.append(float(vertex[i]) + SCREEN_Z)
        new.append(coord)
    return new

def adjustCoordsToRotation(vertices, roll, pitch, yaw):
    new = []
    for vertex in vertices:
        new.append( rotate3D( vertex[0], vertex[1], vertex[2], roll, pitch, yaw ) )
    return new

def flatten(coords):
    # angle of view
    FOV_ANGLE = 46

    x = coords[0]
    y = coords[1]
    z = coords[2]
    angle_radians = (FOV_ANGLE/180) * math.pi
    return (x/(z*math.tan(angle_radians/2)), y/(z*math.tan(angle_radians/2)))
    # return(x/z, y/z)

def rotate3D(x, y, z, roll, pitch, yaw):
    new_x = math.cos(yaw) * math.cos(pitch) * x + \
        (math.cos(yaw) * math.sin(pitch) * math.sin(roll) - math.sin(yaw) * math.cos(roll)) * y + \
        (math.cos(yaw) * math.sin(pitch) * math.cos(roll) + math.sin(yaw) * math.sin(roll))  * z
    new_y = math.sin(yaw) * math.cos(pitch) * x + \
        (math.sin(yaw) * math.sin(pitch) * math.sin(roll) + math.cos(yaw) * math.cos(roll)) * y + \
        (math.sin(yaw) * math.sin(pitch) * math.cos(roll) - math.cos(yaw) * math.sin(roll))  * z
    new_z = -math.sin(pitch) * x + \
        math.cos(pitch) * math.sin(roll) * y + \
        math.cos(pitch) * math.cos(roll) * z
    return([new_x, new_y, new_z])

# def rotateCube():

    
```

```python
# define screen size
WIDTH = 800
HEIGHT = 800

# define screen variables from world 0,0
SCREEN_Z = 15
SCREEN_X = 2
SCREEN_Y = 1

ROLL = 0
PITCH = 0
YAW = 0

DELAY = 1

master = tk.Tk()

canvas = tk.Canvas(master, width=640, height=480)
canvas.pack()


# add screen distance
screen_adjusted_vertices = adjustCoordsToScreen(vertices, SCREEN_X, SCREEN_Y, SCREEN_Z)

# draw each face
lines = []

for face in faces:
    coords_3D = getFaceCoords(face, screen_adjusted_vertices)

    coords_2D = []
    for coords in coords_3D:
        coords_2D.append(flatten(coords))

    x1 = math.ceil(coords_2D[0][0] * WIDTH)
    y1 = math.ceil(coords_2D[0][1] * HEIGHT)
    x2 = math.ceil(coords_2D[1][0] * WIDTH)
    y2 = math.ceil(coords_2D[1][1] * HEIGHT)
    x3 = math.ceil(coords_2D[2][0] * WIDTH)
    y3 = math.ceil(coords_2D[2][1] * HEIGHT)

    # print(x1, y1, x2, y2, x3, y3)
    lines.append(canvas.create_line(x1,y1,x2,y2, fill="green", width=2))
    lines.append(canvas.create_line(x2,y2,x3,y3, fill="green", width=2))
    lines.append(canvas.create_line(x3,y3,x1,y1, fill="green", width=2))

canvas.bind_all('<Key-Right>', right_arrow)  
canvas.bind_all('<Key-Left>', left_arrow) 
canvas.bind_all('<Key-Down>', down_arrow) 
canvas.bind_all('<q>', q)
canvas.bind_all('<e>', e)
canvas.bind_all('<w>', w)
canvas.bind_all('<a>', a)
canvas.bind_all('<s>', s)
canvas.bind_all('<d>', d)

master.mainloop()
```
